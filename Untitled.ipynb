```python
import pandas as pd
df = pd.read_csv('jumbo_beer.csv', index_col=0)
```

```python
amounts = []
for k in range(len(df)):
    name = df.iloc[k]['name']
    amount = None
    try:
        if 'PAK' in df.iloc[k]['id'] or 'TRA' in df.iloc[k]['id'] :
            for i in range(len(name.split(' '))):
                if name.split(' ')[i] == 'x':
                    if 'cl' in name.split(' ')[i+1]:
                        amount = int(name.split(' ')[i-1]) * int(name.split(' ')[i+1].split('cl')[0])
                        break
        elif 'FLS' in df.iloc[k]['id']:
            if 'cl' in name.split(' ')[-1]:
                amount = int(name.split(' ')[-1].split('cl')[0])
        amounts.append(amount)
    except Exception:
        amounts.append(None)
```

```python
df['amounts'] = amounts
df['price per cl'] = df['price']/df['amounts']
```
